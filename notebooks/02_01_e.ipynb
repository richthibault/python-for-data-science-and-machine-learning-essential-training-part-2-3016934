# Simple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from pylab import rcParams

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10, 8
```

```python
rooms = 2*np.random.randn(100,1)+3
rooms[0:10]
```

```python
# so cool we can do these calcs without looping
weights = np.array([10,20,32,18])
heights = np.array([5,6,6,6])
bmi = weights/heights**2
bmi
```

```python
price = 265 + 6*rooms +abs(np.random.randn(100, 1))
price[0:10]
```

```python
plt.plot(rooms,price, 'r^')
plt.xlabel('# of Rooms, 2024 Average')
plt.ylabel('2024 Average Home Price, $1000s USD')
plt.show()
```

```python
X = rooms
y = price

LinReg = LinearRegression()
LinReg.fit(X,y)
print(LinReg.intercept_, LinReg.coef_)
```

*Simple Algebra*
- y = mx + b
- b = intercept = 265.97

*Estimated Coefficients*
- LinReg.coef_ = [5.95] Estimated coefficients for the terms in the linear regression problem.

```python
print(LinReg.score(X,y))
```
